```python
import json
import ast
df =elgibiltiy_df.copy()
def fix_json_column(col):
    def convert(x):
        # If already a dict → dump to JSON
        if isinstance(x, dict):
            return json.dumps(x)

        # Try JSON first
        try:
            return json.dumps(json.loads(x))
        except:
            pass

        # Try Python dict format
        try:
            py_obj = ast.literal_eval(x)
            return json.dumps(py_obj)
        except Exception as e:
            print("BAD JSON:", x)
            raise e

    return col.apply(convert)


df["patient_json"] = fix_json_column(df["patient_json"])
df["eligibility_json"] = fix_json_column(df["eligibility_json"])
```

```python
df
```

```python
def make_example(row):
    instruction = (
        "Decide whether the following patient is eligible for the given clinical trial, "
        "and provide reasoning for your decision."
    )

    input_data = {
        "patient": json.loads(row["patient_json"]),
        "trial_eligibility": json.loads(row["eligibility_json"]),
    }

    output = {
        "is_eligible": bool(row["is_eligible"]),
        "reasoning": row["reasoning"]
    }

    return {
        "instruction": instruction,
        "input": input_data,
        "output": output,
    }

examples = df.apply(make_example, axis=1).tolist()
```

```python

# 2. Your data (you said your list is called `examples`)
# Each item in `examples` looks like your first example:
# {'instruction': '...', 'input': {...}, 'output': {...}}

from datasets import Dataset
import json

# Convert your list into the exact chat format Llama-3.1-Instruct was trained on
def formatting_prompts_func(example):
    messages = [
        {"role": "system", "content": "You are an expert in assessing eligibility for oncology clinical trials. Decide whether the patient is eligible and explain your reasoning step by step."},
        {"role": "user",   "content": example["instruction"] + "\n\nPatient data and trial criteria:\n" + json.dumps(example["input"], indent=2)},
        {"role": "assistant", "content": json.dumps(example["output"], indent=2)}   # we output raw JSON as in your example
    ]
    # Unsloth expects the Llama-3 chat template
    text = "".join([f"<|begin_of_text|><|start_header_id|>{msg['role']}<|end_header_id|>\n\n{msg['content']}<|eot_id|>" for msg in messages]) + "<|start_header_id|>assistant<|end_header_id|>\n\n"
    return {"text": text}

# Build the Hugging Face Dataset
raw_texts = [formatting_prompts_func(ex) for ex in examples]
dataset = Dataset.from_list(raw_texts)

# 3. Load model with Unsloth (4-bit quantized → fits on one 24GB GPU)
from unsloth import FastLanguageModel
import torch

max_seq_length = 4096
dtype = None  # Auto detection
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",  # 4-bit version
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)

# 4. Add LoRA adapters
model = FastLanguageModel.get_peft_model(
    model,
    r=64,                   # rank (64 is good default for 8B)
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                    "gate_proj", "up_proj", "down_proj"],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=3407,
    max_seq_length=max_seq_length,
)
```

```python

# 5. Training arguments
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    dataset_num_proc=2,
    packing=False,  # set to True if you want packing (faster but less readable)
    args=TrainingArguments(
        per_device_train_batch_size=2,      # adjust based on your GPU
        gradient_accumulation_steps=8,
        warmup_steps=2,
        max_steps=10,                      # increase if you have more data
        learning_rate=2e-4,
        fp16=not torch.cuda.is_bf16_supported(),
        bf16=torch.cuda.is_bf16_supported(),
        logging_steps=10,
        output_dir="llama3.1-8b-clinical-eligibility",
        optim="adamw_8bit",
        seed=3407,
        report_to="none",                   # or "wandb" if you want logging
    ),
)
```

```python

# 6. Start training
trainer.train()

```

```python
# 7. Save the final model
model.save_pretrained("llama3.1-8b-clinical-eligibility-final")
tokenizer.save_pretrained("llama3.1-8b-clinical-eligibility-final")
```

```python
from unsloth import FastLanguageModel
import json

# model, tokenizer = FastLanguageModel.from_pretrained(
#     "llama3.1-8b-clinical-eligibility-final",
#     dtype=None,
#     load_in_4bit=True,
# )

FastLanguageModel.for_inference(model)

def is_patient_eligible(patient_dict, trial_dict):
    user_content = f"""Decide whether the following patient is eligible for the given clinical trial, and provide reasoning for your decision.

Patient data and trial criteria:
{json.dumps({"patient": patient_dict, "trial_eligibility": trial_dict}, indent=2)}"""

    messages = [
        {"role": "system", "content": "You are an expert in assessing eligibility for oncology clinical trials. Always respond in strict JSON format: {{\"is_eligible\": true/false, \"reasoning\": \"...\"}}"},
        {"role": "user", "content": user_content}
    ]

    inputs = tokenizer.apply_chat_template(messages, tokenize=True, add_generation_prompt=True, return_tensors="pt").to("cuda")

    outputs = model.generate(inputs, max_new_tokens=512, temperature=0.1, top_p=0.9)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract the assistant's reply
    assistant_reply = response.split("assistant")[-1].strip()
    return json.loads(assistant_reply)   # directly parse the JSON

```
